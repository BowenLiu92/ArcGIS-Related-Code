```python
#New household number calculation: Population - group quarter population:

pop = pd.read_excel('file:///H:\lvpcprojects\Pop%20+%20Emp%20Projections%202018\
/Feb%202019%20Update\TAZ%20based%20results\POP%20result.xlsx', sheet_name = 'LV')

group = pd.read_excel('file:///L:\lvpcgis\BowenLiuTemporary\I-78%20Adams%20Rd%20Point%20of%20Access%20Study\
/ExcelFiles\Aug%20Data%20Update\Household%20Population.xlsx', sheet_name = 'Group Quarter').drop(columns = 'TAZ', axis = 1)
group= group.rename(columns = {'TAZLABEL':'TAZ'})

df = pd.merge(pop, group, on = 'TAZ', how = 'left')


for i in range(2020, 2050, 5):
    df['HH Population ' + str(i)] = df['New' + str(i)] - df[str(i) + ' GPQT POP']
df
```

```python
#Filter out the columns that we need to include in the final table
df.columns
df = df[['TAZ', 'MUNICODE_x', 'NAME', 'COUNTY','HH Population 2020', 'HH Population 2025',
       'HH Population 2030', 'HH Population 2035', 'HH Population 2040',
       'HH Population 2045']]
df.head()
```

```python
#Calculate the household number based on old household number data.

gq_size = pd.read_excel(r'file:///L:\lvpcgis\BowenLiuTemporary\I-78%20Adams%20Rd%20Point%20of%20Access%20Study\
/ExcelFiles\Aug%20Data%20Update\New%20Household%20Number.xlsx', sheet_name = 'New Household Calculation')[['2020 HH Size', '2025 HH Size', '2030 HH Size',
       '2035 HH Size', '2040 HH Size', '2045 HH Size']]

for i in range(2020, 2050, 5):
    df['HH Number ' + str(i)] = df['HH Population ' + str(i)] / gq_size[str(i) + ' HH Size']
df
```

```python
#Save the result to excel file.

folderfilename = input('Give me the file location and also the file name')
writer = pd.ExcelWriter(folderfilename)
df.to_excel(writer, 'Household Number and Pop')
writer.save()
```

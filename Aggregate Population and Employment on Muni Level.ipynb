```python
#Now we can write the function so it can be done for the NC

def reportPE(filename, sheetname):
    df = pd.read_excel(filename, sheet_name = sheetname, header = 2)
    df = df.iloc[:, np.r_[1:5, 5:len(df.columns):2]]
    df = df.groupby(['NAME', 'COUNTY'])['New 2010 POP', 'New 2015 POP',
       'New 2020 POP', 'New 2025 POP', 'New 2030 POP', 'New 2035 POP',
       'New 2040 POP', 'New 2045 POP'].sum().reset_index()
    df = df.append(df.sum(numeric_only=True), ignore_index=True)
    return df

df_NC = reportPE('file:///H:\lvpcprojects\Pop%20+%20Emp%20Projections%202018\Feb%202019%20Update\
/Population%20Growth%20Rate%20and%20Update.xlsx', 'NC POP Update')
```

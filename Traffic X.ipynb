```python
#Now I will do it so that points with different number of records can be differentiated.

dff = pd.read_excel('file:///H:\lvpcprojects\evoLVe\GIS\TrafficX\TrafficX_Updated_Nov16.xlsx', 
                   sheet_name = 'TrafficData for Stack').iloc[:,:6]
dff1 = dff.groupby(['Route', 'ID']).size().reset_index()
dff1.columns = ['Route', 'ID', 'Count']
dff1
dfc3 = dff1[dff1['Count'] == 3]
dfc3
```

```python
#Almost forgot that I need to change the NaN value of T% to string
df2.columns
df2[['Truck_%0', 'Truck_%1', 'Truck_%2', 'Truck_%3', 'Truck_%4']] = df2[['Truck_%0', 'Truck_%1', 'Truck_%2', 'Truck_%3', 'Truck_%4']].fillna('NA')
df2.head()
```

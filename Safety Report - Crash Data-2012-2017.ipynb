```python
#I will do two cycles for the data: 2012 - 2016 and 2013 - 2017

#Using the easy from Stackoverflow

df_c1 = df_m[df_m.CRASH_YEAR != 2012]
df_c1.CRASH_YEAR.unique()

#Doing the same thing and subtotal for cycle 1

df_fulllevel = df_c1.groupby(['MUNICODE','NAME', 'NAMELABEL', 'CRASH_YEAR'])['FATAL_COUNT', 'MAJ_INJ_COUNT', 'MOD_INJ_COUNT', 
                                                                             'MIN_INJ_COUNT', 'BICYCLE_DEATH_COUNT', 'BICYCLE_MAJ_INJ_COUNT',
                                                                            'PED_DEATH_COUNT', 'PED_MAJ_INJ_COUNT'].\
sum()

df_lowerlevel = df_c1.groupby(['MUNICODE','NAME', 'NAMELABEL'])['FATAL_COUNT', 'MAJ_INJ_COUNT', 'MOD_INJ_COUNT', 
                                                                'MIN_INJ_COUNT', 'BICYCLE_DEATH_COUNT', 'BICYCLE_MAJ_INJ_COUNT',
                                                               'PED_DEATH_COUNT', 'PED_MAJ_INJ_COUNT'].\
sum()
df_pr1 = df_lowerlevel.set_index(np.array(['Total'] * len(df_lowerlevel)), append=True) 

cycle1 = pd.concat([df_fulllevel, df_pr1]).sort_index(level=[0,1])
cycle1
```

```python
#Doing the same thing for the 2nd cycle

df_c2 = df_m[df_m.CRASH_YEAR != 2017]
df_c2.CRASH_YEAR.unique()

#Doing the same thing and subtotal for cycle 1

df_fulllevel2 = df_c2.groupby(['MUNICODE','NAME', 'NAMELABEL', 'CRASH_YEAR'])['FATAL_COUNT', 'MAJ_INJ_COUNT', 'MOD_INJ_COUNT', 
                                                                             'MIN_INJ_COUNT', 'BICYCLE_DEATH_COUNT', 'BICYCLE_MAJ_INJ_COUNT',
                                                                            'PED_DEATH_COUNT', 'PED_MAJ_INJ_COUNT'].\
sum()

df_lowerlevel2 = df_c2.groupby(['MUNICODE','NAME', 'NAMELABEL'])['FATAL_COUNT', 'MAJ_INJ_COUNT', 'MOD_INJ_COUNT', 
                                                                'MIN_INJ_COUNT', 'BICYCLE_DEATH_COUNT', 'BICYCLE_MAJ_INJ_COUNT',
                                                               'PED_DEATH_COUNT', 'PED_MAJ_INJ_COUNT'].\
sum()
df_pr2 = df_lowerlevel2.set_index(np.array(['Total'] * len(df_lowerlevel2)), append=True) 

cycle2 = pd.concat([df_fulllevel2, df_pr2]).sort_index(level= [0, 1])

cycle2
```

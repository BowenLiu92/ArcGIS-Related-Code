```python
#Now I need to merge all the excel sheets using the functions I wrote.

#The function
def GetEach(LastBit):
    FilePath = r'file:///H:\lvpcprojects\Transportation\GIS\Crash_data\Excel_files\COMMVEH_' + LastBit + '.csv'
    df = pd.read_csv(FilePath)
    df = df[['CRN', 'VEH_CONFIG_CD']]
    return df

#I will use for loop to join all the dataframes created using the function above

r = []
for i in range(2, 8):
    MyInput = '201' + str(i) + '_Northampton'
    r.append(GetEach(MyInput)) 

df_NC = pd.concat(r)

p = []
for i in range(2, 8):
    MyInput = '201' + str(i) + '_Lehigh'
    p.append(GetEach(MyInput)) 

df_LC = pd.concat(p)

df_commveh = pd.concat([df_LC, df_NC]).reset_index()

df_commveh.head()
df_commveh.shape
```

```python
#Filter the rows based on the vehicle type
df = df_commveh[(df_commveh['VEH_CONFIG_CD'] >= 6)&(df_commveh['VEH_CONFIG_CD'] <= 9)]

df.shape

#Now I need to merge the 2 df (muni and fatality etc df and commveh df)

df_wtruck = pd.merge(df_m, df.set_index('CRN'), left_on= 'CRN', right_index= True)

df_wtruck.head()

df_wtruck.shape
```

```python
#Now I will divide this into two dataframes: for cycle 1 and cycle 2.

#Cycle 1
df_c1 = df_wtruck[df_wtruck.CRASH_YEAR != 2017]

#Cycle 2
df_c2 = df_wtruck[df_wtruck.CRASH_YEAR != 2012]

#For each cycle I need to calculate the sum&average of the fatality, average of the major injuries.

df1 = df_c1.groupby('CRASH_YEAR').agg({'FATAL_COUNT': [np.sum], 'MAJ_INJ_COUNT': np.sum}).rename(columns = {"FATAL_COUNT":"Fatal Sum", "MAJ_INJ_COUNT":"MAJ_INJ_Sum"})

df1.columns.get_level_values(1)
df1.columns = df1.columns.droplevel(1)

df1 = df1.reset_index()
df1

#Make the process above a function so that it can be applied to cycle 2 easily

def TruckGroupbyYear(x):
    x = x.groupby('CRASH_YEAR').agg({'FATAL_COUNT': [np.sum], 'MAJ_INJ_COUNT': np.sum}).rename(columns = {"FATAL_COUNT":"Fatal Sum", "MAJ_INJ_COUNT":"MAJ_INJ_Sum"})
    x.columns = x.columns.droplevel(1)
    x = x.reset_index()
    return x

TruckGroupbyYear(df_c2)
df2 = TruckGroupbyYear(df_c2)
```

```python
#Add sum row to both result dataframes
df1.loc['Total'] = df1.sum()
df1.at['Total', 'CRASH_YEAR'] = None
df1

df2.loc['Total'] = df2.sum()
df2.at['Total', 'CRASH_YEAR'] = None
df2
```

```python
#Groupby municipalities and see the fatality and injury numbers of each municipality
def TruckGroupbyMuni(x):
    x = x.groupby(['NAME', 'NAMELABEL']).agg({'FATAL_COUNT': [np.sum], 'MAJ_INJ_COUNT': np.sum}).rename(columns = {"FATAL_COUNT":"Fatal Sum", "MAJ_INJ_COUNT":"MAJ_INJ_Sum"})
    x.columns = x.columns.droplevel(1)
    x = x.reset_index()
    return x
TruckGroupbyMuni(df_c1)
TruckGroupbyMuni(df_c2)
```

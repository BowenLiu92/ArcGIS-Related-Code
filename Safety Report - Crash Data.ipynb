```python
#Try Merge and get the Muniname

df1 = df
df_muniraw = pd.read_excel('file:///L:\lvpcgis\Transportation\
/Walk_BikeLV\Crash%20data\MunicipalityCrashData\CrashDataAnalysis\MuniExcel.xlsx')
df_muniraw.columns
```

```python
#Select the columns that I need
dfm = df_muniraw[['COUNTY', 'MUNICODE', 'NAME', 'NAMELABEL', 'CountyName', 'PDOT_MCode']]
dfm.head()
dfm
```

```python
#Try rename just the COUNTY column
dfm = dfm.rename(columns ={df.columns[0]: 'LVPC_COUNTY'})
dfm
```

```python
#Try joining the two dataframes for now
df_m = pd.merge(df1, dfm.set_index('PDOT_MCode'), left_on = 'MUNICIPALITY', right_index = True )
df_m.head()
df_m.columns
```

```python
#I will groupby the counties for this one.
df_m.groupby(['COUNTY', 'LVPC_COUNTY', 'CountyName', 'CRASH_YEAR'])['BICYCLE_DEATH_COUNT', 'BICYCLE_MAJ_INJ_COUNT'].sum()
df_m.groupby(['COUNTY', 'LVPC_COUNTY', 'CountyName', 'CRASH_YEAR'])[['BICYCLE_DEATH_COUNT', 'BICYCLE_MAJ_INJ_COUNT']].\
sum().reset_index()
```

```python
#Run and groupby municipalities.
df_e1 = df_m.groupby(['MUNICODE','NAME', 'NAMELABEL', 'CRASH_YEAR'])['FATAL_COUNT', 'MAJ_INJ_COUNT', 'MOD_INJ_COUNT', 'MIN_INJ_COUNT'].\
sum()
df_e1
```

```python
writer = pd.ExcelWriter('L:\lvpcgis\Transportation\Walk_BikeLV\Crash data\MunicipalityCrashData\
/CrashDataAnalysis\MuniPreliminarySum.xlsx')
df_e1.to_excel(writer,'Muni')
writer.save()
```

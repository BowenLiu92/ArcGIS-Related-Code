```python
#Apparently to save them into one Excel workbook is not that easy. 

#Create a small function to filter each sheet
def cleanTab(v):
    df = pd.read_excel('file:///H:\lvpcprojects\FutureLV\ScenarioPlanning\LV_LivingWorkingFlow\
/LV_Worker_FlowII\MuniLevelData_All_Inflow.xlsx', sheet_name = v)
    df = df[(df.Municipalities.str.contains('Lehigh, PA')) | (df.Municipalities.str.contains('Northampton, PA'))]
    return df
d ={}
for name in l1:
    d[name] = cleanTab(name)
```

```python


def save_xls(dict_df, path):
    
    writer = pd.ExcelWriter(path)
    for key in dict_df:
        dict_df[key].to_excel(writer, '%s' % key)

    writer.save()
```
